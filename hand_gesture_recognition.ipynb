```python
import os
import shutil
from sklearn.model_selection import train_test_split

def split_and_organize_dataset(base_path, train_ratio=0.7, val_ratio=0.2, test_ratio=0.1):
    # Paths for training, validation, and test sets
    train_path = os.path.join(base_path, 'train')
    val_path = os.path.join(base_path, 'validation')
    test_path = os.path.join(base_path, 'test')

    # Create directories if they don't exist
    for path in [train_path, val_path, test_path]:
        os.makedirs(path, exist_ok=True)

    for folder in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder)

        # Skip train/validation/test folders to avoid re-processing
        if folder in ['train', 'validation', 'test']:
            continue

        if os.path.isdir(folder_path):
            print(f"Processing folder: {folder}")
            images = []
            labels = []

            for subfolder in os.listdir(folder_path):
                subfolder_path = os.path.join(folder_path, subfolder)
                label = subfolder  # Class label

                for file_name in os.listdir(subfolder_path):
                    file_path = os.path.join(subfolder_path, file_name)
                    images.append(file_path)
                    labels.append(label)

            # Debug: print the number of images found
            print(f"Found {len(images)} images in folder: {folder}")

            if len(images) == 0:
                print(f"No images found in folder {folder}. Skipping...")
                continue

            # Split the dataset
            train_images, temp_images, train_labels, temp_labels = train_test_split(
                images, labels, train_size=train_ratio, stratify=labels)

            val_images, test_images, val_labels, test_labels = train_test_split(
                temp_images, temp_labels, train_size=val_ratio/(val_ratio + test_ratio), stratify=temp_labels)

            # Move files to respective directories
            def move_files(image_list, label_list, set_path):
                for image, label in zip(image_list, label_list):
                    label_path = os.path.join(set_path, label)
                    os.makedirs(label_path, exist_ok=True)

                    # Construct the destination file path
                    dest_file_path = os.path.join(label_path, os.path.basename(image))

                    # If the file already exists, handle it (e.g., rename or skip)
                    if os.path.exists(dest_file_path):
                        print(f"File {dest_file_path} already exists. Skipping or renaming...")
                        continue
                        # Alternative: Rename the file
                        # new_dest_file_path = dest_file_path.replace('.png', '_dup.png')
                        # shutil.move(image, new_dest_file_path)
                    else:
                        shutil.move(image, dest_file_path)

            move_files(train_images, train_labels, train_path)
            move_files(val_images, val_labels, val_path)
            move_files(test_images, test_labels, test_path)

    print("Dataset splitting and organization complete.")

base_path = '/content/leapGestRecog/leapGestRecog'
split_and_organize_dataset(base_path)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping, CSVLogger

# Define paths and parameters
base_path = '/content/leapGestRecog/leapGestRecog'
target_size = (128, 128)  # Consider reducing to (64, 64) if still facing issues
batch_size = 32  # Reduced batch size to reduce memory usage
epochs = 5

# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    base_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    base_path,
    target_size=target_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Model definition using VGG16
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(target_size[0], target_size[1], 3))
base_model.trainable = False

model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(len(train_generator.class_indices), activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Callbacks
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=3, verbose=1)
lr_reduction = ReduceLROnPlateau(monitor='loss', patience=3, verbose=1, factor=0.5)
csv_logger = CSVLogger('training_log.csv', append=True)  # Log training process

# Model training
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs,
    callbacks=[checkpoint, early_stopping, lr_reduction, csv_logger]
)

# Print final accuracy
print(f"Final training accuracy: {history.history['accuracy'][-1]:.4f}")
print(f"Final validation accuracy: {history.history['val_accuracy'][-1]:.4f}")
```

```python
# Save the trained model in the native Keras format
model.save('gesture_recognition_model.keras')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Make predictions on the validation set
y_pred = model.predict(validation_generator, steps=validation_generator.samples // validation_generator.batch_size + 1)
y_pred_classes = np.argmax(y_pred, axis=1)

# Compute confusion matrix
y_true = validation_generator.classes
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=validation_generator.class_indices.keys(),
            yticklabels=validation_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report

# Get class counts
val_labels = pd.Series(validation_generator.classes)
class_counts = val_labels.value_counts()
print(class_counts)

# Display a sample of predictions
print("Sample of predictions:", np.unique(y_pred_classes))
print("Sample of true labels:", np.unique(y_true))

# Define class names based on the validation generator
class_names = list(validation_generator.class_indices.keys())

# Generate the classification report
report = classification_report(
    y_true,
    y_pred_classes,
    target_names=class_names,
    zero_division=0,
    labels=list(validation_generator.class_indices.values())
)
print(report)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to display images with predictions
def display_predictions(generator, model, num_images=5):
    plt.figure(figsize=(15, 10))

    # Fetch a batch of images and labels
    images, labels = next(generator)

    for i in range(num_images):
        # Predict the class
        pred = model.predict(np.expand_dims(images[i], axis=0))
        pred_class = np.argmax(pred, axis=1)[0]
        true_class = np.argmax(labels[i], axis=0)

        # Display the image with the prediction and true label
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i])
        plt.title(f"True: {list(generator.class_indices.keys())[true_class]}\nPred: {list(generator.class_indices.keys())[pred_class]}")
        plt.axis('off')

    plt.show()

# Use the function with your validation generator
display_predictions(validation_generator, model)
```
